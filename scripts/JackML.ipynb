## Step One: Data Preprocessing

Import the dataset 'covabdab_all.csv' using the Python 'pandas' library and then start preliminary data processing by removing unecessary columns from the dataset. Unecessary columns can be identified by determining whether it's value would have a substantial impact on the overall prediction of the model. After removing irrelevant columns, Further preprocess the data by determining which values are not needed or if certain value may affect the overall prediction. i.e. selecting human patients only, removing conflicting values, etc. Identify missing data values 'NaN' or 'ND' and determine whether to remove them or add a supplementary value. 


```python
import pandas

#import dataset
df = pandas.read_csv("covabdab_all.csv")

#prelimary data processing
df = df.drop(columns=['Name', 'Structures', 'ABB Homology Model (if no structure)', 'Sources', 'Date Added', 'Last Updated', 'Update Description', 'Notes/Following Up?'])

#selecting value of importance
df = df[df['Origin'].str.contains('Human') == True]
df = df[((df['Neutralising Vs'].notna() == True) | (df['Not Neutralising Vs'].notna() == True)) &
        ~((df['Neutralising Vs'].notna() == True) & (df['Not Neutralising Vs'].notna() == True))]

#removing empty values
df = df[df['VHorVHH'].notna() == True]

df.head()
```

```python
df.to_csv("test.csv", index=False)
```
